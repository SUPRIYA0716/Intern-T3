### Importing Libraries

```python
#To work with dataframe
import pandas as pd

#To work with numerical operations
import numpy as np

#To visualize the data
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

### Reading Data

```python
# Reading datas
datas = pd.read_csv('Samplesuperstore.csv')
datas.head(7)
```

## DATA PRE-PROCESSING

```python
datas.drop_duplicates(subset=None , keep ='first' , inplace=True)
```

```python
datas.duplicated().sum()
```

```python
df=datas.corr()
```

```python
sns.heatmap(df , annot=True ,cmap='Greens')
```

```python
datas = datas.drop(['Postal Code'],axis=1)
```

## DATA VISUALIZATION

```python
sns.pairplot(datas , hue='Ship Mode')
```

```python
datas['Ship Mode'].value_counts()
```

```python
sns.countplot(x=datas['Ship Mode'])
```

```python
datas['Segment'].value_counts()
```

```python
sns.pairplot(datas, hue='Segment')
```

```python
sns.countplot(x='Segment' , data = datas)
```

```python
sns.countplot(x='Category' , data=datas)
```

```python
datas['Sub-Category'].value_counts()
```

```python
plt.figure(figsize=(10,10))
datas['Sub-Category'].value_counts().plot.pie()
plt.legend()
plt.show()
```

```python
datas['State'].value_counts()
```

```python
plt.figure(figsize=(10,10))
datas['State'].value_counts().plot.pie()
#plt.legend()
plt.show()
```

```python
datas.hist(figsize=(10,10))
plt.show()
```

```python
plt.figure(figsize=(6,6))
datas['Region'].value_counts().plot.pie()
plt.legend()
plt.show()
```

```python
fig,ax=plt.subplots(figsize=(7,7))
ax.scatter(datas['Sales'],datas['Profit'])
plt.xlabel("sales")
plt.ylabel("profit")
plt.show()
```

```python
sns.regplot(x='Discount',y='Profit',label='Profit',data=datas)
plt.legend()
plt.show()
```

```python
sns.regplot(x='Quantity',y='Profit',label='Profit',data=datas)
plt.legend()
plt.show()
```

```python
datas.groupby('Segment')[['Profit' , 'Sales']].sum().plot.bar(color=['violet','green'],figsize=(7,4))
plt.ylabel('profit or loss sales')
plt.show()
```

```python
states=datas['State'].value_counts().nlargest(25)
states
```

```python
plt.figure(figsize=(7,7))
plt.title('Sales in Each Region')
sns.barplot(x='Region',y='Sales',data=datas,hue='Segment',order=datas['Region'].value_counts().index)
plt.show()
```
